**VẼ BIỂU ĐỒ THÔNG QUA THƯ VIỆN SEABORN**

Mục đích chính của Seaborn là giúp việc tạo ra các biểu đồ thống kê đẹp mắt, hấp dẫn và giàu thông tin trở nên đơn giản và nhanh chóng hơn

```python
#Biểu đồ tương quan (Relational Plots)
import seaborn as sns
import matplotlib.pyplot as plt

# Tải dữ liệu mẫu "tips" (tiền boa trong nhà hàng)
tips_df = sns.load_dataset("tips")

# Vẽ biểu đồ(Quan trong nhất là tham số hue)
sns.scatterplot(
    data=tips_df,        # DataFrame chứa dữ liệu
    x="total_bill",      # Tên cột cho trục X
    y="tip",             # Tên cột cho trục Y
    hue="sex",           # Phân loại màu sắc theo cột 'sex'
    style="time"         # Phân loại kiểu điểm theo cột 'time'
)

plt.title("Tiền Tip so với Tổng Hóa đơn")
plt.show()
```

```python
# Biểu đồ Tần suất (history chart)
penguins_df = sns.load_dataset("penguins")

# Vẽ biểu đồ
sns.histplot(
    data=penguins_df,
    x="flipper_length_mm",  # Tên cột cần xem phân bố
    bins=20,                # Chia dữ liệu thành 20 (cột)
    kde=True                # Vẽ thêm đường ước tính mật độ (KDE)
)

plt.title("Phân bố độ dài Vây của Chim cánh cụt")
plt.show()
```

```python
# Biểu đồ hộp
tips_df = sns.load_dataset("tips")

# Vẽ biểu đồ
sns.boxplot(
    data=tips_df,
    x="day",        # Biến phân loại cho trục X
    y="total_bill", # Biến số để xem phân bố
    hue="smoker"    # Chia nhỏ mỗi hộp theo người hút thuốc/không
)

plt.title("Phân bố Tổng hóa đơn theo Ngày và Người hút thuốc")
plt.show()
```

```python
# Biểu đồ đếm
titanic_df = sns.load_dataset("titanic")

# Vẽ biểu đồ
sns.countplot(
    data=titanic_df,
    x="class",       # Tên cột cần đếm
    hue="survived"   # Chia mỗi cột thành các nhóm nhỏ hơn
)

plt.title("Số lượng Hành khách theo Hạng vé và Tình trạng Sống sót")
plt.show()
```

```python
#Biểu đồ nhiệt
penguins_df = sns.load_dataset("penguins")

# 1. Chọn các cột số
numeric_cols = penguins_df.select_dtypes(include=['number'])

# 2. Tính ma trận tương quan
corr_matrix = numeric_cols.corr()

# 3. Vẽ biểu đồ nhiệt
sns.heatmap(
    corr_matrix,   # Dữ liệu ma trận
    annot=True,    # Hiển thị giá trị số trong mỗi ô
    cmap="coolwarm", # Bảng màu
    fmt=".2f"      # Định dạng số (2 chữ số sau dấu phẩy)
)

plt.title("Ma trận Tương quan giữa các Đặc tính của Chim cánh cụt")
plt.show()
```
